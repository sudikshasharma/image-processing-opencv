```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('../../Data/DATA/00-puppy.jpg')
img2 = cv2.imread('../../Data/DATA/watermark_no_copy.png')
plt.imshow(img1)
```

```python
# BLENDING (using addWeighted()=> it asks for both images with same dimensions)
# So, step 1 : resize images and make them of same dimension/shape
img1_new = cv2.resize(img1, (900,900))
img2_new = cv2.resize(img2, (900,900))
```

```python
# To blend image, one method in cv2 is to use weights => blendedPixel = a.pixel_img1+b.pixel_img2+g
# Here, alpha (a), beta (b) and gamma(g) are weights to img1's pixel, img2's pixel and adjustor respectively.
# More is the weight, more intense is that image's intensity in blended image (eg ig a=0.1,b = 0.9, img one is almost invisible nad img2 is almost of full intensity)

blended_img = cv2.addWeighted(img1_new,0.5,img2_new,0.5,0.8)
plt.imshow(blended_img)
```

```python
# PASTE one image on other (NOT BLENDING)
#Slice the image(background image) over which we want to paste another image and get array size of the to be pasted image. Assign pixel of that chunk in backgroud image as the values of front image 

# img1 = background image
# img2 = foreground (pasted) image

img1[:img2.shape[0],:img2.shape[1]] = img2
```
