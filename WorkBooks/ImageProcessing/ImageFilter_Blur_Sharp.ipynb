```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Defining method to load image
def load_img():
    img = cv2.imread("../../Data//Data/bricks.jpg").astype(np.float32())/255
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
# Expanding displayed image
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
img = load_img()
display_img(img)
```

```python
# gamma correction is increasing/decreasing brightness of image (gamma is brightness)
# gamma = 1 (normal intensity), gamma <1 (increased brightness), gamma>1 (decreased brightness)
# as gamma decreases, brightness of image increases
# Power function is used for gamma correction
gamma = 0.3
img_gamma = np.power(img, gamma)
display_img(img_gamma)
```

```python
gamma = 1.4
img_gamma = np.power(img, gamma)
display_img(img_gamma)
```
