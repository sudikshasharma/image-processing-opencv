```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
# Create 3d array of all 0 and data type = int
img_ar = np.zeros(shape = (400,400,3),dtype = np.int16)
img_ar.shape
```

```python
plt.imshow(img_ar)
```

```python
# draw rec with parameters : image(gets changed), left top coord, right bottom coord, color, thickness
cv2.rectangle(img_ar,(100,100),(250,200),color=(0,255,255),thickness=5)
plt.imshow(img_ar)
```

```python
cv2.rectangle(img_ar,(200,200),(300,300),(0,255,0),8)
plt.imshow(img_ar)
```

```python
# Better(optional) to name the parameters as well for understanding. (Fix names as defined by methods,can't change them)
cv2.circle(img = img_ar,center = (250,200),color = (0,250,0),radius = 20,thickness=3)
plt.imshow(img_ar)
```

```python
# To make any shape be filled, put thickness to any -ve number (here, its -2, but can be any -ve number)
cv2.circle(img = img_ar,center = (100,300),color = (0,250,0),radius = 20,thickness=-2)
plt.imshow(img_ar)
```

```python
cv2.line(img=img_ar,pt1=(0,400),pt2=(300,150),color=(255,0,0),thickness=2)
plt.imshow(img_ar)
# will give error if thickness of line is -ve
```

```python
#Put text in image
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img=img_ar,text="Hey!",org=(0,400),fontFace = font,fontScale=4,color=(255,0,0),thickness=3)
plt.imshow(img_ar)
```

```python
img_poly = np.zeros(shape=(500,500,3),dtype = np.int16)
plt.imshow(img_poly)
```

```python
# define points of polygon in array
pts = np.array([[100,100],[100,200],[300,400],[400,200]],dtype=np.int32)
# make it a 3d array(add another coordinate for points)
pts = pts.reshape(-1,1,2)
```

```python
# syntax to draw a custom polygon, pts should be list =>[pts], isClosed tells whether polygon is closed
cv2.polylines(img_poly,[pts],isClosed = True, color = (255,0,255),thickness=3)
plt.imshow(img_poly)
```
